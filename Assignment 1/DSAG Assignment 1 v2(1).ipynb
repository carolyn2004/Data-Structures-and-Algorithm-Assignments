### a. Briefly describe two main differences between Big O notation and Big Omega  notation. (6 marks)

The first difference between between Big O notation and Big Omega notation is that Big O represents the upper bound of the alogorithm while the Big Omega represents the lower bound of the algorithm. T(n) is Ω(f(n)) iff some constants c and n0 T(n)≥cf(n) for all n≥n0 - T(n) is Ω(f(n)) basically means that f(n) describes the lower bound for T(n) for large n.

The second difference between Big O notation and Big Omega notation is that Big O is used to describe the worst case running time for an algorithm. But, Big Omega notation, on the other hand, is used to describe the best case running time for a given algorithm. Since Big Omega represents the lower bound of the algorithm, Big Omega only cares about the lowest amount of work (faster) and is therefore used to describe the best case running time of the algorithm. On the other hand, Big O notation represents the upper bound of the algorithm. Big O represents the upper bound of the algorithm. It is associated with the highest amount of work (slower). Thus, Big O is used to describe the worst case running time for an algorithm.

### Write a function named add_big(first_num, second_num). This function adds two "big number" together. "big number" here is represented by a list of digits. For e.g. The number 135678965 is represented by [1,3,5,6,7,8,9,6,5]. If you call the function with first_num=[8,2,4,5,5,5] and second_num=[4,5,6,5,5,5], the function will add 824555 to 456555, and return  [1, 2, 8, 1, 1, 1, 0] as a result.
### Note that the two numbers do not need to be of the same size and the function should not modify the lists sent by the caller.

### The marking scheme for Q4 to Q6 is given at the end of Q6.

```python
def add_big(first_num,second_num):
    # joining the digits back to form a number as string
    first_str = ""
    for num in first_num: 
        first_str += str(num)
    second_str = ""
    for num1 in second_num: 
        second_str += str(num1)
    # typecasting the number in string back to integer
    first_int = int(first_str)
    second_int = int(second_str)
    # adding both numbers together
    sum_int = first_int + second_int
    # separate digits back into the list which will be output 
    output_list = []
    for num2 in str(sum_int):
        output_list.append(num2)
    return output_list
    
firstlist=[8,2,4,5,5,5]
secondlist=[4,5,6,5,5,5]
add_big (firstlist, secondlist)
```

### Write a program to prompt user to enter a time in HH:MM:SS format. The program displays the time and it simulates the running of a clock. You are only allowed to use the sleep function in the time module to implement this simulated clock program. A sample run of the program is capture in the video here: [demo](https://studenttpedu-my.sharepoint.com/:v:/g/personal/kokyau_tp_edu_sg/EXBJ8EGoqPBGvuc8oAn0KRoBKHN3a1B_Y16KZSglZOFWWg?e=9lWQja)

```python
time = input("Please enter time hr:min:sec ").split(":") #allow user to input time
a_list = [] #create an empty list
 
for i in time: 
    a_list.append(int(i)) 
 
 
hours = a_list[0] 
minutes = a_list[1] 
seconds = a_list[2] 
 
import time #import time module
 
 
while True: #create a while loop so time will keep printing
    if hours > 23: #if user inputs more than 23 hours it will be considered invalid
        print("Invalid time entered")
        break # to ensure statement will not printed more than once
    elif minutes > 59: #if user inputs any value more than 59 minutes it will be considered invalid
        print("Invalid time entered")
        break # to ensure statement will not be printed more than once
    elif seconds > 59: #if user inputs any value more than 59 seconds it will be considered invalid
        print("Invalid time entered")
        break # to ensure statement will not be printed more than once
    else:
        print(f"{hours:02d}: {minutes:02d}: {seconds:02d}", end="\r") # ensure only 2 digits of each variable is being printed
        time.sleep(1)
        seconds=seconds +1 #increment the time every one second 
 
        if seconds == 60: #when seconds reaches 60, seconds will become 0 and 1 will be added to minutes
            seconds = 0
            minutes = minutes + 1
        elif minutes == 60: #when minutes reaches 60, minutes will become 0, and 1 will be added to hours
            minutes = 0
            hours = hours + 1
        elif hours == 24: #when hours reaches 24, the hours will be 0 and stimulated clock programme will reset
            hours = 0
```
