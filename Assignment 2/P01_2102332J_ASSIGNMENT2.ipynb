```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~“‘'''
def remove_punctuation(word): 
    word_only = ""
    for ch in word: 
        if ch not in punctuations:
            word_only += ch
    return (word_only)

#this function has a for loop which goes through each character and check if theres a punctuation 
#if ch is not a punctuation in the list provided, it will add on to the string called words_only


stopwords = [] #empty list
with open("stop_words_english.txt", "r", encoding = "UTF-8") as my_stopwords: 
    for word in my_stopwords:
        word = remove_punctuation(word) #remove all punctuations from stopwords using remove_punctuation
        word = word.rsplit() #splits a string into a list
        if word[0] not in stopwords:
            stopwords += word


bookwords = [] #empty list
with open("book_03.txt", "r", encoding = "utf-8-sig") as my_book: 
    for line in my_book:
        line = line.rstrip("\n")
        line = line.rsplit()
        for word in line:
            word = remove_punctuation(word)
            word = word.lower()
            if  (word not in stopwords) and (word.isalpha() == True): 
                bookwords.append(word)
                
#remove all punctuations and numbers from words 
#isalpha() checks whether the string consists of alphabetic characters only
#we need to convert the words to all lowercase because if for example there are 2 of the same words - STOP and stop, 
#python will process these 2 words as different although they are the same word
#this will affect our count for frequncy and it will no longer be accurate
                
bookwords_freq_dict = {} #empty dictionary
for word in bookwords:
    if word not in bookwords_freq_dict.keys(): 
        bookwords_freq_dict[word] = 1
    else:
        bookwords_freq_dict[word] += 1
        
#if the word is not in the dictionary, the value remains the same
#else if the word is already in the dictionary we increase the value by 1

        
bookwords_freq_list = []
for key, value in bookwords_freq_dict.items(): 
    bookwords_freq_list.append(value)
bookwords_freq_list.sort(reverse = True) 

#we want to get the top 10 most frequent words
#sort() method sorts the list ascending by default

top_ten_freq = bookwords_freq_list[0:15]
top_ten_freq_unique = []
for freq in top_ten_freq: 
    if freq not in top_ten_freq_unique:
        top_ten_freq_unique.append(freq)
        
#use slicing to get top 10 most frequenct words in a new list

bookwords_freq_dict_swapped = {}
for key, value in bookwords_freq_dict.items():
    if value not in bookwords_freq_dict_swapped: 
        bookwords_freq_dict_swapped[value] = []
        bookwords_freq_dict_swapped[value].append(key)
    else:
        bookwords_freq_dict_swapped[value].append(key) 
        
#we swap the key and value to get the output as word, followed by frequency value
        
for freq in top_ten_freq_unique:
    words_list = bookwords_freq_dict_swapped[freq]
    for word in words_list:
        print (word, freq)
#note to teacher: please restart kernel before running this code if needed
```

### A computer programmer is to generate the following sequence of numbers:
###		2	5	9	19	37  	.  .  .
###	Write a recursive function to generate the first 10 numbers of the sequence.



```python
def recur_seq(n):
    if n==:
        return 0
#stopping/terminating condition

    elif n%2==0:
        return recur_seq(n-1)*2+1
#if n%2==0 it indicates the n is an even number because even numbers can divide by 2 without giving a remainder
    else:
        return recur_seq(n-1)*2-1
#if the previous condition is not met it shows that there is a remainder when n is divided by 2, proving that n is an odd number
    
#there is an alternating pattern between the number sequence hence 2 conditions are specified

#for example, (2)*2+1 = 5 where n=2(even number +1)
#where as, (5)*2-1 = 9 where n=3 (odd number +1)
#(n-1) refers to the previous term

for i in range(0, 11):
    print(recur_seq(i), end=' ')
#use a for loop to print the first 10 numbers of the sequence
#use end=' ' so that all the numbers will be printed in one row/line rather than in one column


        
        
```
